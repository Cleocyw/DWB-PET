## an example of uploading whole-body image

```python
part = np.load('wb-time1/output1-450.npy')
for i in range(7):
    body_part = np.load('wb-time1/output{}-450.npy'.format(i+2))
    part = np.append(part,body_part,axis=1)

#rotate the image
import scipy.ndimage as ndimage
angle = 90
wb = ndimage.rotate(part, angle, reshape=True)

plt.imshow(wb,cmap = 'hot')
plt.colorbar()
plt.clim(0,17000)
plt.savefig('denoised-wb14.png')
```
