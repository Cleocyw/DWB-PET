```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
label = np.load('label-1.npy')
output_mean = np.load('output1-mean.npy')
output_blurry = np.load('output1-blurry.npy')
output_noise = np.load('output1-noise.npy')
```

```python
ax = plt.subplot()
diff_mean=plt.imshow((label-output_mean)/label,cmap = 'bwr')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(diff_mean, cax=cax)
plt.clim(-0.5,0.5)
plt.savefig('diff_mean.png')
```

```python
ax = plt.subplot()
diff_blurry=plt.imshow((label-output_blurry)/label,cmap = 'bwr')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(diff_blurry, cax=cax)
plt.clim(-0.5,0.5)
plt.savefig('diff_blurry.png')
```

```python
ax = plt.subplot()
diff_noise=plt.imshow((label-output_noise)/label,cmap = 'bwr')
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(diff_noise, cax=cax)
plt.clim(-0.5,0.5)
plt.savefig('diff_noise.png')
```
